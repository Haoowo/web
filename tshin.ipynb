```python
#線上申辦:申請帳單  

import requests
from bs4 import BeautifulSoup
import pandas as pd
url = f"https://www.taishinbank.com.tw/TSB/personal/credit/online-apply/e-bill/"
 
# 發送 GET 請求獲取 HTML 內容
headers = {'user-agent': '"Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_2) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/47.0.2526.106 Safari/537.36'}
response = requests.get(url, headers=headers)

# 使用 BeautifulSoup 解析 HTML 內容
soup = BeautifulSoup(response.text, 'html.parser')
formdata = []

# 找到所有包含資訊的<div>元素
ser_div = soup.find_all('div', class_='editor-txt editor-ul-li editor-ol-li editor-table htmlcontent')
for ser_divs in ser_div:
    paragraph_texts = []
    for paragraph in ser_divs:
        if paragraph.find('tr'):
            continue
        else:
            paragraphs = ser_divs.find_all('span')  
            paragraph_texts.append(paragraph.text)  
        # Print the extracted paragraph texts
    for text in paragraph_texts:
        print(text,"\n")

    # Initialize lists to store table data
    

    # Loop through each row in the table
    for row in ser_divs.find_all('tr'):
        # Initialize list to store row data
        row_data = []
        # Loop through each cell in the row
        for cell in row.find_all('td'):
        # Append cell content to row data
            row_data.append(cell.get_text(strip=True))
     # Append row data to the main data list
        formdata.append(row_data)

    # Print the extracted data
    for row in formdata:
        print(row)
        
    # 找到所有包含資訊的<div>元素
ser_div = soup.find_all('div', class_='ts-comp-24')
for ser_divs in ser_div:
    content2= ser_divs.find(class_='desc').text.strip()
    print(content2)
```
